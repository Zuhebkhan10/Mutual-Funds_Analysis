```python
#scaling the model
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
rank = df[[

           'Returns_1Yr',
           'Returns_3Yr',
           'Returns_5Yr',
           'Expense_Ratio',
           'Risk_Level',
           'Alpha',
           'Sd',
           'Beta',
           'Sharpe',
           'Rating'
]]
rank = scaler.fit_transform(rank)
rank
```

```python
#ranking score, Before using this fromula we can perform the scaling amd transfrom it
# Higher expense means more charges. Investors prefer lower costs.
# Higher Beta means higher market volatility. More volatility = More risk
# Higher SD means higher fluctuations in returns. Investors generally prefer stable funds
# Returns = Most important factor → 30%
# Sharpe Ratio = Important for risk-adjusted performance → 20%
# Alpha = Fund manager skill → 20%
# Rating = Overall quality → 15%
# Risk and Cost factors → Remaining 15%
df['Score'] = (
    0.30 * df['Returns_5Yr'] +0.20 * df['Sharpe'] +0.20 * df['Alpha'] +
    0.15 * df['Rating'] -0.05 * df['Expense_Ratio'] -0.05 * df['Beta'] -0.05 * df['Sd']
    )
print(df['Score'])
```

```python
# check the feature engineering steps
features = [
    'Expense_Ratio',
    'Returns_1Yr',
    'Returns_3Yr',
    'Returns_5Yr',
    'Sharpe',
    'Sortino',
    'Alpha',
    'Beta'
]
X = df[features].copy()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
scaled_df = pd.DataFrame(X_scaled,columns=features)

scaled_df.head()
```

```python
df['fund_score'] = (0.10 * scaled_df['Returns_1Yr']+ 0.15 * scaled_df['Returns_3Yr']+ 0.25 * scaled_df['Returns_5Yr']
    + 0.15 * scaled_df['Sharpe'] + 0.10 * scaled_df['Sortino']+ 0.15 * scaled_df['Alpha']
    - 0.05 * scaled_df['Expense_Ratio'] - 0.05 * scaled_df['Beta']
)
```
