```python
#check the outliers by using the z score test if the value is z score > 3 --> outlier, z score < 3 --> outlier
from scipy.stats import zscore
df['z_score']=zscore(df['Min_Sip'])
outliers=df[(df['z_score']>3| (df['z_score']<-3))].sum()
print(outliers)
```
