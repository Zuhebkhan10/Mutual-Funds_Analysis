```python
# Check the outlier detection
q1=df['Min_Sip'].quantile(0.25)
q3=df['Min_Sip'].quantile(0.75)
iqr=q3-q1

lower_limit=150-(1.5*iqr)
lower_limit
```

```python
upper_limit=1000+(1.5*iqr)
upper_limit

# There is no Outliers, if there  max and min value in range of upper_limit and lower_limit ther is no outliers.  
```

```python
# check the distribution of data
sns.histplot(df['Min_Sip'])
plt.show()
```

```python
#check the outliers by using the z score test if the value is z score > 3 --> outlier, z score < 3 --> outlier
from scipy.stats import zscore
df['z_score']=zscore(df['Min_Sip'])
outliers=df[(df['z_score']>3| (df['z_score']<-3))].sum()
print(outliers)
```
